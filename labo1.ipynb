```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

Z = range(0, 20)
F = np.linspace(0, 2 * np.pi, 100)

n = sp.symbols('n', Float=True)
x1, x2, x3 = sp.symbols('x1 x2 x3', cls=sp.Function)

x1 = sp.sin(0.1 * sp.pi * n + sp.pi / 4)
x2 = (-1)**n
x3 = sp.KroneckerDelta(n, 10)
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,3,1)  
plt.stem(Z, [x1.subs(n, i) for i in Z], label='x1')
plt.subplot(1,3,2)
plt.stem(Z, [x2.subs(n, i) for i in Z], label='x2')
plt.subplot(1,3,3)
plt.stem(Z, [x3.subs(n, i) for i in Z], label='x3')
plt.show()
```

```python
zx1 = [float(x1.subs(n, i)) for i in Z]
zx2 = [float(x2.subs(n, i)) for i in Z]
zx3 = [float(x3.subs(n, i)) for i in Z]    

fftx1 = np.fft.fft(zx1)
fftx2 = np.fft.fft(zx2)
fftx3 = np.fft.fft(zx3)

shifted_fftx1 = np.fft.fftshift(fftx1)
shifted_fftx2 = np.fft.fftshift(fftx2)
shifted_fftx3 = np.fft.fftshift(fftx3)

plt.figure(figsize=(16,4))
plt.subplot(1,3,1)  
plt.stem(Z, shifted_fftx1, label='x1')
plt.subplot(1,3,2)
plt.stem(Z, shifted_fftx2, label='x2')
plt.subplot(1,3,3)
plt.stem(Z, shifted_fftx3, label='x3')
plt.show()
```
