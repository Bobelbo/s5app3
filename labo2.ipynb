```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp


x = range(0,100)

Fc = 2000
Fe = 16000

n, N = sp.symbols('n N', Float=True)
h, k= sp.symbols('h k', cls=sp.Function)

k = 2 * N * (Fc/Fe) + 1
h = 1/N * (sp.sin(sp.pi * n * (k/N))/sp.sin(sp.pi * (n/N)))
```

```python
h
```

```python
plt.figure(figsize=(12,4))
max = 64
for N_v, c in [(64, "green"), (32, "yellow"), (16, "red")]:
    H = h.subs(N, N_v)
    r = range(0,N_v)
    arr = [H.subs(n, i) for i in r]
    arr[0] = k.subs(N,N_v) / N_v
    
    arr = np.fft.fftshift(arr)
    
    arr = arr[1:]
    
    plt.stem(r[1:], arr, linefmt=c)
    
    
```
